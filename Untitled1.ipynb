```python
# --- Step 1: Install Libraries ---
!pip install pandas numpy scikit-learn matplotlib seaborn

# --- Step 2: Import Libraries ---
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns

# --- Step 3: Simulate Student Data ---
np.random.seed(42)
students = ['S1','S2','S3','S4','S5']
topics = ['Algebra','Geometry','Physics','Chemistry','Biology']

data = []
for student in students:
    for topic in topics:
        score = np.random.randint(40,100)  # Random scores 40-100
        time_spent = np.random.randint(20,60)  # Minutes spent
        # NEW FEATURE: Include a difficulty level (1=Easy, 2=Medium, 3=Hard)
        difficulty = np.random.choice([1, 2, 3])
        data.append([student, topic, score, time_spent, difficulty])

df = pd.DataFrame(data, columns=['StudentID','Topic','Score','TimeSpent', 'Difficulty'])

# Encode topic for ML
df['Topic_encoded'] = df['Topic'].astype('category').cat.codes

# Define performance: Weak=0 (<60), Medium=1 (60-79), Strong=2 (80+)
df['Performance'] = df['Score'].apply(lambda x: 0 if x<60 else (1 if x<80 else 2))
print("Sample Data:")
print(df.head())

# --- Step 4: Split Features and Labels (Now including Difficulty) ---
X = df[['Topic_encoded','TimeSpent', 'Difficulty']] # Added 'Difficulty' as a feature
y = df['Performance']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# --- Step 5: Train Random Forest Model ---
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Evaluate
y_pred = model.predict(X_test)
print("\n--- Model Evaluation ---")
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))

# --- Step 6: Predict Weak Topics for a New Student ---
# Simulate a new student's study time per topic and current difficulty
new_student_data = pd.DataFrame({
    'Topic_encoded': [0, 1, 2, 3, 4],  # Algebra, Geometry, Physics, Chemistry, Biology
    'TimeSpent': [25, 35, 30, 20, 40],
    'Difficulty': [2, 1, 3, 2, 1] # Current assumed difficulty level for the student
})
new_student_topics = ['Algebra', 'Geometry', 'Physics', 'Chemistry', 'Biology']

pred = model.predict(new_student_data)
# NEW FEATURE: Predict the probability distribution (confidence)
pred_proba = model.predict_proba(new_student_data)

pred_labels = ['Weak','Medium','Strong']
pred_text = [pred_labels[i] for i in pred]
print("\n--- Predicted Performance ---")
print("Predicted performance per topic:", pred_text)

# ----------------------------------------------------------------------
# --- NEW FEATURE 1: Confidence Score & Adaptive Difficulty ---
# ----------------------------------------------------------------------

print("\n--- AI Confidence & Next Steps ---")
difficulty_map = {1: 'Easy', 2: 'Medium', 3: 'Hard'}

for i, topic in enumerate(new_student_topics):
    confidence = np.max(pred_proba[i]) * 100
    current_diff = new_student_data.loc[i, 'Difficulty']

    # Adaptive Difficulty Logic
    next_action = ""
    if pred[i] == 2 and confidence > 80: # Strong and High Confidence
        next_diff = min(current_diff + 1, 3) # Max difficulty is 3 (Hard)
        next_action = f"Level Up! Recommended Difficulty: {difficulty_map[next_diff]}."
    elif pred[i] == 0: # Weak
        next_action = "Focus on basics."
    else: # Medium or Strong but low confidence
        next_action = f"Continue practice at current difficulty ({difficulty_map[current_diff]})."

    print(f"| {topic}: Performance={pred_text[i]} | Confidence={confidence:.1f}% | Action: {next_action}")


# ----------------------------------------------------------------------
# --- NEW FEATURE 2: NLP-Based Dynamic Question Generator Function ---
# (Carried over from the previous update)
# ----------------------------------------------------------------------

def generate_question(topic, difficulty_level):
    """Generates a unique practice question based on topic and difficulty."""

    templates = {
        1: f"Easy: What is the primary function of a **{topic}** concept?",
        2: f"Medium: Solve this typical **{topic}** problem: [X]",
        3: f"Hard: Analyze and contrast two advanced principles of **{topic}** using a real-world scenario."
    }

    return templates.get(difficulty_level, f"Practice problem for {topic}: Review {difficulty_map[difficulty_level]}.")


# --- Step 7: Dynamic Question Recommendation ---
print("\n--- Dynamic Question Recommendation for Weak/Medium Topics ---")

for i, topic in enumerate(new_student_topics):
    current_diff = new_student_data.loc[i, 'Difficulty']

    if pred[i] == 0:  # Weak
        # Suggest an easier question for a weak topic
        recommended_diff = max(1, current_diff - 1)
        question = generate_question(topic, recommended_diff)
        print(f"üõë Recommended for {topic} (Weak): Dynamic Question (Difficulty {recommended_diff}): {question}")
    elif pred[i] == 1: # Medium
        # Suggest a question at current difficulty
        question = generate_question(topic, current_diff)
        print(f"üî∂ Recommended for {topic} (Medium): Dynamic Question (Difficulty {current_diff}): {question}")


# ----------------------------------------------------------------------
# --- NEW FEATURE 3: Gamification Metrics ---
# ----------------------------------------------------------------------

print("\n--- Gamification Metrics (Simulated) ---")
student_xp = 5500
current_level = int(student_xp / 1000)
daily_streak = 7
topic_mastery = {t: df[df['Topic'] == t]['Score'].mean() for t in topics}

print(f"üåü Student Level: {current_level} (XP: {student_xp})")
print(f"üî• Daily Streak: {daily_streak} days!")
print("üèÜ Topic Mastery Scores:")
for t, score in topic_mastery.items():
    badge = 'ü•á' if score >= 80 else ('ü•à' if score >= 60 else 'ü•â')
    print(f"  - {t}: {badge} {score:.1f}")


# --- Step 8: Visualize Student Progress ---
print("\n--- Student Performance Visualization ---")
df_visual = df.pivot(index='StudentID', columns='Topic', values='Score')
plt.figure(figsize=(10,6))
sns.heatmap(df_visual, annot=True, cmap='YlGnBu', fmt="d", linewidths=.5, cbar_kws={'label': 'Score'})
plt.title("Student Performance Heatmap (80+ is Strong)")
plt.show()
# --- Step 1: Install Libraries ---
!pip install pandas numpy scikit-learn matplotlib seaborn

# --- Step 2: Import Libraries ---
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns

# --- Step 3: Simulate Student Data ---
np.random.seed(42)
students = ['S1','S2','S3','S4','S5']
topics = ['Algebra','Geometry','Physics','Chemistry','Biology']

data = []
for student in students:
    for topic in topics:
        score = np.random.randint(40,100)  # Random scores 40-100
        time_spent = np.random.randint(20,60)  # Minutes spent
        # NEW FEATURE: Include a difficulty level (1=Easy, 2=Medium, 3=Hard)
        difficulty = np.random.choice([1, 2, 3])
        data.append([student, topic, score, time_spent, difficulty])

df = pd.DataFrame(data, columns=['StudentID','Topic','Score','TimeSpent', 'Difficulty'])

# Encode topic for ML
df['Topic_encoded'] = df['Topic'].astype('category').cat.codes

# Define performance: Weak=0 (<60), Medium=1 (60-79), Strong=2 (80+)
df['Performance'] = df['Score'].apply(lambda x: 0 if x<60 else (1 if x<80 else 2))
print("Sample Data:")
print(df.head())

# --- Step 4: Split Features and Labels (Now including Difficulty) ---
X = df[['Topic_encoded','TimeSpent', 'Difficulty']] # Added 'Difficulty' as a feature
y = df['Performance']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# --- Step 5: Train Random Forest Model ---
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Evaluate
y_pred = model.predict(X_test)
print("\n--- Model Evaluation ---")
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))

# --- Step 6: Predict Weak Topics for a New Student ---
# Simulate a new student's study time per topic and current difficulty
new_student_data = pd.DataFrame({
    'Topic_encoded': [0, 1, 2, 3, 4],  # Algebra, Geometry, Physics, Chemistry, Biology
    'TimeSpent': [25, 35, 30, 20, 40],
    'Difficulty': [2, 1, 3, 2, 1] # Current assumed difficulty level for the student
})
new_student_topics = ['Algebra', 'Geometry', 'Physics', 'Chemistry', 'Biology']

pred = model.predict(new_student_data)
# NEW FEATURE: Predict the probability distribution (confidence)
pred_proba = model.predict_proba(new_student_data)

pred_labels = ['Weak','Medium','Strong']
pred_text = [pred_labels[i] for i in pred]
print("\n--- Predicted Performance ---")
print("Predicted performance per topic:", pred_text)

# ----------------------------------------------------------------------
# --- NEW FEATURE 1: Confidence Score & Adaptive Difficulty ---
# ----------------------------------------------------------------------

print("\n--- AI Confidence & Next Steps ---")
difficulty_map = {1: 'Easy', 2: 'Medium', 3: 'Hard'}

for i, topic in enumerate(new_student_topics):
    confidence = np.max(pred_proba[i]) * 100
    current_diff = new_student_data.loc[i, 'Difficulty']

    # Adaptive Difficulty Logic
    next_action = ""
    if pred[i] == 2 and confidence > 80: # Strong and High Confidence
        next_diff = min(current_diff + 1, 3) # Max difficulty is 3 (Hard)
        next_action = f"Level Up! Recommended Difficulty: {difficulty_map[next_diff]}."
    elif pred[i] == 0: # Weak
        next_action = "Focus on basics."
    else: # Medium or Strong but low confidence
        next_action = f"Continue practice at current difficulty ({difficulty_map[current_diff]})."

    print(f"| {topic}: Performance={pred_text[i]} | Confidence={confidence:.1f}% | Action: {next_action}")


# ----------------------------------------------------------------------
# --- NEW FEATURE 2: NLP-Based Dynamic Question Generator Function ---
# (Carried over from the previous update)
# ----------------------------------------------------------------------

def generate_question(topic, difficulty_level):
    """Generates a unique practice question based on topic and difficulty."""

    templates = {
        1: f"Easy: What is the primary function of a **{topic}** concept?",
        2: f"Medium: Solve this typical **{topic}** problem: [X]",
        3: f"Hard: Analyze and contrast two advanced principles of **{topic}** using a real-world scenario."
    }

    return templates.get(difficulty_level, f"Practice problem for {topic}: Review {difficulty_map[difficulty_level]}.")


# --- Step 7: Dynamic Question Recommendation ---
print("\n--- Dynamic Question Recommendation for Weak/Medium Topics ---")

for i, topic in enumerate(new_student_topics):
    current_diff = new_student_data.loc[i, 'Difficulty']

    if pred[i] == 0:  # Weak
        # Suggest an easier question for a weak topic
        recommended_diff = max(1, current_diff - 1)
        question = generate_question(topic, recommended_diff)
        print(f"üõë Recommended for {topic} (Weak): Dynamic Question (Difficulty {recommended_diff}): {question}")
    elif pred[i] == 1: # Medium
        # Suggest a question at current difficulty
        question = generate_question(topic, current_diff)
        print(f"üî∂ Recommended for {topic} (Medium): Dynamic Question (Difficulty {current_diff}): {question}")


# ----------------------------------------------------------------------
# --- NEW FEATURE 3: Gamification Metrics ---
# ----------------------------------------------------------------------

print("\n--- Gamification Metrics (Simulated) ---")
student_xp = 5500
current_level = int(student_xp / 1000)
daily_streak = 7
topic_mastery = {t: df[df['Topic'] == t]['Score'].mean() for t in topics}

print(f"üåü Student Level: {current_level} (XP: {student_xp})")
print(f"üî• Daily Streak: {daily_streak} days!")
print("üèÜ Topic Mastery Scores:")
for t, score in topic_mastery.items():
    badge = 'ü•á' if score >= 80 else ('ü•à' if score >= 60 else 'ü•â')
    print(f"  - {t}: {badge} {score:.1f}")


# --- Step 8: Visualize Student Progress ---
print("\n--- Student Performance Visualization ---")
df_visual = df.pivot(index='StudentID', columns='Topic', values='Score')
plt.figure(figsize=(10,6))
sns.heatmap(df_visual, annot=True, cmap='YlGnBu', fmt="d", linewidths=.5, cbar_kws={'label': 'Score'})
plt.title("Student Performance Heatmap (80+ is Strong)")
plt.show()
```

```python
# --- Step 1: Install Libraries ---
!pip install pandas numpy scikit-learn matplotlib seaborn

# --- Step 2: Import Libraries ---
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, timedelta

# --- Step 3: Simulate Student Data (with temporal/session data) ---
np.random.seed(42)
students = ['S1','S2','S3','S4','S5']
topics = ['Algebra','Geometry','Physics','Chemistry','Biology']
sessions = 5 # Simulate 5 learning sessions over time

data = []
start_date = datetime.now()

for student in students:
    for topic in topics:
        mastery_level = np.random.uniform(0.1, 0.9) # Initial BKT mastery (latent)

        for session in range(1, sessions + 1):
            date = start_date + timedelta(days=session * np.random.randint(1, 3)) # Spaced sessions

            # Score influenced by initial mastery, time since last session (forgetting), and random noise
            forget_factor = 1 - (0.1 * (session - 1)) # Simple linear forgetting simulation
            score_base = mastery_level * 100 * forget_factor
            score = int(np.clip(score_base + np.random.randint(-15, 15), 40, 100))

            time_spent = np.random.randint(20,60)
            difficulty = np.random.choice([1, 2, 3])

            # Update mastery based on score (simplified BKT update)
            if score >= 80: mastery_level = min(1.0, mastery_level + 0.15)
            elif score < 60: mastery_level = max(0.1, mastery_level - 0.10)

            data.append([student, topic, date, session, score, time_spent, difficulty, mastery_level])

df = pd.DataFrame(data, columns=['StudentID','Topic','Date','Session','Score','TimeSpent','Difficulty','LatentMastery'])

# Encode topic for ML
df['Topic_encoded'] = df['Topic'].astype('category').cat.codes
df['Performance'] = df['Score'].apply(lambda x: 0 if x<60 else (1 if x<80 else 2))
print("Sample Data (First 3 sessions for one student/topic):")
print(df[df['StudentID']=='S1'].head(3))

# --- Step 4: Prepare Data for Prediction (Using only the LATEST session data for the model) ---
df_latest = df.sort_values('Date').drop_duplicates(['StudentID', 'Topic'], keep='last').reset_index(drop=True)

X = df_latest[['Topic_encoded','TimeSpent', 'Difficulty']]
y = df_latest['Performance']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# --- Step 5: Train Random Forest Model ---
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# --- Step 6: Predict for a New Student (using simulated latest data) ---
new_student_data = pd.DataFrame({
    'Topic_encoded': [0, 1, 2, 3, 4],
    'TimeSpent': [25, 35, 30, 20, 40],
    'Difficulty': [2, 1, 3, 2, 1]
})
new_student_topics = ['Algebra', 'Geometry', 'Physics', 'Chemistry', 'Biology']

pred = model.predict(new_student_data)
pred_proba = model.predict_proba(new_student_data)
pred_labels = ['Weak','Medium','Strong']
pred_text = [pred_labels[i] for i in pred]

print("\n--- Predicted Performance (Latest Session) ---")
print("Predicted performance per topic:", pred_text)

# ----------------------------------------------------------------------
# --- NEW ADVANCED FEATURE 1: SIMULATED BAYESIAN KNOWLEDGE TRACING (BKT) ---
# --- BKT models the student's mastery level and forgetting curve.
# ----------------------------------------------------------------------

# Simulate a sophisticated BKT output based on the latent mastery score
bkt_output = df_latest[df_latest['StudentID']=='S1'][['Topic', 'LatentMastery']].set_index('Topic')
bkt_output['Forgetting_Rate'] = np.random.uniform(0.05, 0.35, len(bkt_output)) # Simulate a calculated BKT forgetting rate

print("\n--- Knowledge Tracing Analysis (Student S1) ---")
for topic in new_student_topics:
    mastery = bkt_output.loc[topic, 'LatentMastery']
    forget = bkt_output.loc[topic, 'Forgetting_Rate']

    status = 'Mastered' if mastery > 0.8 else ('Learning' if mastery > 0.5 else 'Struggling')
    print(f"| {topic}: Mastery={mastery:.2f} ({status}) | Forgetting Rate={forget:.2f}")

# ----------------------------------------------------------------------
# --- NEW ADVANCED FEATURE 2: SPACED REPETITION SCHEDULING ---
# --- Based on BKT forgetting curve, recommend the next study date.
# ----------------------------------------------------------------------

print("\n--- Adaptive Spaced Repetition Schedule ---")
for topic in new_student_topics:
    # Get the simulated BKT forgetting rate
    forget_rate = bkt_output.loc[topic, 'Forgetting_Rate']

    # Schedule interval: High mastery/Low forgetting -> longer interval
    # Low mastery/High forgetting -> shorter interval

    # Days until next review (e.g., base 7 days - (forget_rate * 20 days))
    days_to_review = int(np.clip(7 - (forget_rate * 10), 2, 14))
    review_date = datetime.now() + timedelta(days=days_to_review)

    print(f"| {topic}: Review in {days_to_review} days. Schedule: {review_date.strftime('%Y-%m-%d')}")

# ----------------------------------------------------------------------
# --- NLP Dynamic Question Generator (Updated with Difficulty) ---
# ----------------------------------------------------------------------

def generate_question(topic, difficulty_level):
    templates = {
        1: f"[{topic} | Easy]: Define the core concept and give one simple example.",
        2: f"[{topic} | Medium]: Solve this typical problem, showing all intermediate steps.",
        3: f"[{topic} | Hard]: Contrast this concept with a related topic and predict an outcome in a novel scenario."
    }
    return templates.get(difficulty_level, f"Practice problem for {topic}: Review concepts.")

# --- Step 7: Dynamic Question Recommendation & Adaptive Difficulty ---
print("\n--- Dynamic Question Recommendation for Weak/Medium Topics ---")

for i, topic in enumerate(new_student_topics):
    current_diff = new_student_data.loc[i, 'Difficulty']

    # Adaptive Difficulty Logic - Level Up/Down
    if pred[i] == 0:  # Weak
        recommended_diff = max(1, current_diff - 1)
        action = "REMEDIATION"
    elif pred[i] == 2 and np.max(pred_proba[i]) > 0.8: # Strong and High Confidence
        recommended_diff = min(3, current_diff + 1)
        action = "CHALLENGE"
    else:
        recommended_diff = current_diff
        action = "PRACTICE"

    question = generate_question(topic, recommended_diff)
    print(f"[{action} | D={recommended_diff}] {topic}: {question}")


# --- Step 8: Visualize Temporal Progress (New Visualization) ---
print("\n--- Student Temporal Progress Visualization ---")

# Select one student for the longitudinal plot
student_id = 'S1'
df_s1 = df[df['StudentID'] == student_id]

plt.figure(figsize=(12, 6))
sns.lineplot(data=df_s1, x='Session', y='Score', hue='Topic', marker='o')
plt.title(f"Score Progression Over Time for {student_id}")
plt.xlabel("Learning Session")
plt.ylabel("Score (%)")
plt.grid(True)
plt.show()
# --- Step 1: Install Libraries ---
!pip install pandas numpy scikit-learn matplotlib seaborn

# --- Step 2: Import Libraries ---
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, timedelta

# --- Step 3: Simulate Student Data (with temporal/session data) ---
np.random.seed(42)
students = ['S1','S2','S3','S4','S5']
topics = ['Algebra','Geometry','Physics','Chemistry','Biology']
sessions = 5 # Simulate 5 learning sessions over time

data = []
start_date = datetime.now()

for student in students:
    for topic in topics:
        mastery_level = np.random.uniform(0.1, 0.9) # Initial BKT mastery (latent)

        for session in range(1, sessions + 1):
            date = start_date + timedelta(days=session * np.random.randint(1, 3)) # Spaced sessions

            # Score influenced by initial mastery, time since last session (forgetting), and random noise
            forget_factor = 1 - (0.1 * (session - 1)) # Simple linear forgetting simulation
            score_base = mastery_level * 100 * forget_factor
            score = int(np.clip(score_base + np.random.randint(-15, 15), 40, 100))

            time_spent = np.random.randint(20,60)
            difficulty = np.random.choice([1, 2, 3])

            # Update mastery based on score (simplified BKT update)
            if score >= 80: mastery_level = min(1.0, mastery_level + 0.15)
            elif score < 60: mastery_level = max(0.1, mastery_level - 0.10)

            data.append([student, topic, date, session, score, time_spent, difficulty, mastery_level])

df = pd.DataFrame(data, columns=['StudentID','Topic','Date','Session','Score','TimeSpent','Difficulty','LatentMastery'])

# Encode topic for ML
df['Topic_encoded'] = df['Topic'].astype('category').cat.codes
df['Performance'] = df['Score'].apply(lambda x: 0 if x<60 else (1 if x<80 else 2))
print("Sample Data (First 3 sessions for one student/topic):")
print(df[df['StudentID']=='S1'].head(3))

# --- Step 4: Prepare Data for Prediction (Using only the LATEST session data for the model) ---
df_latest = df.sort_values('Date').drop_duplicates(['StudentID', 'Topic'], keep='last').reset_index(drop=True)

X = df_latest[['Topic_encoded','TimeSpent', 'Difficulty']]
y = df_latest['Performance']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# --- Step 5: Train Random Forest Model ---
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# --- Step 6: Predict for a New Student (using simulated latest data) ---
new_student_data = pd.DataFrame({
    'Topic_encoded': [0, 1, 2, 3, 4],
    'TimeSpent': [25, 35, 30, 20, 40],
    'Difficulty': [2, 1, 3, 2, 1]
})
new_student_topics = ['Algebra', 'Geometry', 'Physics', 'Chemistry', 'Biology']

pred = model.predict(new_student_data)
pred_proba = model.predict_proba(new_student_data)
pred_labels = ['Weak','Medium','Strong']
pred_text = [pred_labels[i] for i in pred]

print("\n--- Predicted Performance (Latest Session) ---")
print("Predicted performance per topic:", pred_text)

# ----------------------------------------------------------------------
# --- NEW ADVANCED FEATURE 1: SIMULATED BAYESIAN KNOWLEDGE TRACING (BKT) ---
# --- BKT models the student's mastery level and forgetting curve.
# ----------------------------------------------------------------------

# Simulate a sophisticated BKT output based on the latent mastery score
bkt_output = df_latest[df_latest['StudentID']=='S1'][['Topic', 'LatentMastery']].set_index('Topic')
bkt_output['Forgetting_Rate'] = np.random.uniform(0.05, 0.35, len(bkt_output)) # Simulate a calculated BKT forgetting rate

print("\n--- Knowledge Tracing Analysis (Student S1) ---")
for topic in new_student_topics:
    mastery = bkt_output.loc[topic, 'LatentMastery']
    forget = bkt_output.loc[topic, 'Forgetting_Rate']

    status = 'Mastered' if mastery > 0.8 else ('Learning' if mastery > 0.5 else 'Struggling')
    print(f"| {topic}: Mastery={mastery:.2f} ({status}) | Forgetting Rate={forget:.2f}")

# ----------------------------------------------------------------------
# --- NEW ADVANCED FEATURE 2: SPACED REPETITION SCHEDULING ---
# --- Based on BKT forgetting curve, recommend the next study date.
# ----------------------------------------------------------------------

print("\n--- Adaptive Spaced Repetition Schedule ---")
for topic in new_student_topics:
    # Get the simulated BKT forgetting rate
    forget_rate = bkt_output.loc[topic, 'Forgetting_Rate']

    # Schedule interval: High mastery/Low forgetting -> longer interval
    # Low mastery/High forgetting -> shorter interval

    # Days until next review (e.g., base 7 days - (forget_rate * 20 days))
    days_to_review = int(np.clip(7 - (forget_rate * 10), 2, 14))
    review_date = datetime.now() + timedelta(days=days_to_review)

    print(f"| {topic}: Review in {days_to_review} days. Schedule: {review_date.strftime('%Y-%m-%d')}")

# ----------------------------------------------------------------------
# --- NLP Dynamic Question Generator (Updated with Difficulty) ---
# ----------------------------------------------------------------------

def generate_question(topic, difficulty_level):
    templates = {
        1: f"[{topic} | Easy]: Define the core concept and give one simple example.",
        2: f"[{topic} | Medium]: Solve this typical problem, showing all intermediate steps.",
        3: f"[{topic} | Hard]: Contrast this concept with a related topic and predict an outcome in a novel scenario."
    }
    return templates.get(difficulty_level, f"Practice problem for {topic}: Review concepts.")

# --- Step 7: Dynamic Question Recommendation & Adaptive Difficulty ---
print("\n--- Dynamic Question Recommendation for Weak/Medium Topics ---")

for i, topic in enumerate(new_student_topics):
    current_diff = new_student_data.loc[i, 'Difficulty']

    # Adaptive Difficulty Logic - Level Up/Down
    if pred[i] == 0:  # Weak
        recommended_diff = max(1, current_diff - 1)
        action = "REMEDIATION"
    elif pred[i] == 2 and np.max(pred_proba[i]) > 0.8: # Strong and High Confidence
        recommended_diff = min(3, current_diff + 1)
        action = "CHALLENGE"
    else:
        recommended_diff = current_diff
        action = "PRACTICE"

    question = generate_question(topic, recommended_diff)
    print(f"[{action} | D={recommended_diff}] {topic}: {question}")


# --- Step 8: Visualize Temporal Progress (New Visualization) ---
print("\n--- Student Temporal Progress Visualization ---")

# Select one student for the longitudinal plot
student_id = 'S1'
df_s1 = df[df['StudentID'] == student_id]

plt.figure(figsize=(12, 6))
sns.lineplot(data=df_s1, x='Session', y='Score', hue='Topic', marker='o')
plt.title(f"Score Progression Over Time for {student_id}")
plt.xlabel("Learning Session")
plt.ylabel("Score (%)")
plt.grid(True)
plt.show()
```
